The data we are intrested in residdes in the columns `label` and `LocTime_utc`.

Then, we will pivot those columns to get the dat into a usefull format.  

```python
# Label column as the forllowing format "CarData - variable_name - value"
split_label_col = raw_ts["label"].str.split(" - ", expand=True)
raw_ts["value"] = split_label_col[2]
raw_ts["variable"] = split_label_col[1]


ts:DF = pd.pivot_table(raw_ts, values="value", index="LocTime_utc", columns="variable", aggfunc="first")
ts
```
