### Data extraction

```python
raw_tss = get_raw_tss("tesla")
```

## Analysis

```python
var_wise_density = raw_tss.count().div(raw_tss.shape[0])
display(var_wise_density)
var_wise_density.to_csv("data_cache/var_wise_density.csv")
```

```python
vin_raw_tss_groups = raw_tss.groupby("vin")
vehicle_var_wise_density = vin_raw_tss_groups.count().div(vin_raw_tss_groups.size(), axis=0)
display(vehicle_var_wise_density)
vehicle_var_wise_density.to_csv("data_cache/vehicle_var_wise_density.csv")
```
