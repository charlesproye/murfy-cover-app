### Data extraction

```python
tss = ProcessedTimeSeries("volvo-cars")
```

```python
raw_tss = get_raw_tss("volvo-cars")
raw_tss
```

```python
raw_tss_to_plot = (
    raw_tss
    .rename(columns={
        "diagnostics.distance_since_reset":"distance_since_reset",
        "usage.average_speed":"average_speed",
        "usage.electric_consumption_average":"average_electric_consumption",
    })
    .eval("distance_since_reset = distance_since_reset.ffill()")
    .eval("average_speed = average_speed.ffill()")
)
```

```python
px.scatter(
    raw_tss_to_plot,
    x="distance_since_reset",
    y="average_electric_consumption",
    color="vin"
).show()
```

```python

plt_3d_df(
    raw_tss_to_plot.query("average_electric_consumption < 50 & average_electric_consumption > 10 & distance_since_reset > 100 & average_speed > 0"),
    x="distance_since_reset",
    y="average_speed",
    z="average_electric_consumption",
    color=""
)
```
