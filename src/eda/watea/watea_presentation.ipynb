```python
processed_cluster["floored_temperature"] = floor_to(processed_cluster["temperature"], 5)
dist_to_plot = (
    processed_cluster
    .query("temperature < 30 & temperature > 0")
    .groupby(["soc", "floored_temperature"])[["energy_added"]]
    .median()
    .reset_index()
    .sort_values(by=["floored_temperature", "soc"], ascending=[False, True])
)
fig = px.line(
    dist_to_plot,
    x="soc",
    y="energy_added",
    color="floored_temperature",
    color_discrete_sequence=px.colors.sequential.Rainbow,
)

# fig.update_layout(
#         title = 'Number of Items by Year',
#     yaxis_title='Number of items',
#     xaxis_title='Year',
#     paper_bgcolor = 'white', # set background color as white
#     plot_bgcolor = 'white',  # set background color as white

#     xaxis = dict(
#         showline = True,
#         linecolor = 'rgb(102, 102, 102)',
#         tickfont_color = 'rgb(102, 102, 102)',
#         showticklabels = True,
#         dtick = 10, # interval of each tick in x-axis. In this case: 1600, 1610, 1620, etc.
#         ticks = 'outside',
#         tickcolor = 'rgb(102, 102, 102)',
#     ),
#      yaxis = dict(
#         showline = True,
#         linecolor = 'rgb(102, 102, 102)',
#         tickfont_color = 'rgb(102, 102, 102)',
#         showticklabels = True,
#         dtick = 5, # interval of each tick in y-axis. In this case: 0, 5, 10, 15, etc.
#         ticks = 'outside',
#         tickcolor = 'rgb(102, 102, 102)',
#     ),

# )
# fig.update_layout(hovermode='x unified')

fig.show()
```
