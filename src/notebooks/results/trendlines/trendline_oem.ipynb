# Oem trendline

Show trendline compute on oem.

```python
from core.sql_utils import *
import plotly.graph_objects as go
import plotly.express as px
from core.numpy_utils import numpy_safe_eval
import numpy as np
```

```python
with get_connection() as connection:
    query = """
    SELECT o.oem_name, o.trendline, o.trendline_max, o.trendline_min
    FROM oem o
    """
    df = pd.read_sql(query, connection)
```

```python
for oem in df.dropna()['oem_name'].unique():
    df_temp = df[df['oem_name']==oem]
    trendline = df_temp['trendline'].values[0]
    trendline_max = df_temp['trendline_max'].values[0]
    trendline_min = df_temp['trendline_min'].values[0]
    print(trendline)
    print(trendline_max)
    print(trendline_min)
    fig = go.Figure()
    x_sorted = np.array([i*1000 for i in range(200)])
    fig.add_trace(go.Scatter(
        x=x_sorted,
        y=numpy_safe_eval(expression=trendline['trendline'], x=x_sorted),
        mode='lines',
        line=dict(color='red'),
        name='Fit'
    ))

    fig.add_trace(go.Scatter(
        x=x_sorted,
        y=numpy_safe_eval(expression=trendline_max['trendline'], x=x_sorted),
        mode='lines',
        line=dict(color='green'),
        name='Upper'
    ))

    fig.add_trace(go.Scatter(
        x=x_sorted,
        y=numpy_safe_eval(expression=trendline_min['trendline'], x=x_sorted),
        mode='lines',
        line=dict(color='green'),
        name='Lower'
    ))
    fig.update_layout(
        title=f"{oem}"
    )
    fig.show()
```
